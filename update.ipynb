```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

model = AutoModelForCausalLM.from_pretrained(
    "Qwen/Qwen1.5-MoE-A2.7B-Chat",
    torch_dtype=torch.bfloat16,
    device_map="auto",
    attn_implementation="sdpa"
)

model.eval()
```

```python
tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen1.5-MoE-A2.7B-Chat")

prompt= "What are the names of the busiest streets in cities with a population exceeding 20 million?"

messages = [
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": prompt}
]
```

```python
text = tokenizer.apply_chat_template(
    messages,
    tokenize=False,
    add_generation_prompt=True
)
model_inputs = tokenizer([text], return_tensors="pt").to("cuda")
```

```python
with torch.no_grad():
  outputs = model(
      input_ids = model_inputs["input_ids"],
      output_router_logits=True
  )
```

```python
router_logits =outputs.router_logits
router_logits
```

```python
import torch

router_logits[0].shape
logits_layer0=router_logits[0]

top_k = model.config.top_k

# https://docs.pytorch.org/docs/stable/generated/torch.topk.html
topk_scores, topk_indices = torch.topk(logits_layer0, k=top_k, dim=-1)
#print(topk_scores)
#print(topk_indices)

# Flatten all assignments
assigned_experts = topk_indices.flatten()
# assigned_experts shape: (seq_len * k)

num_experts = logits_layer0.shape[-1]
counts = torch.bincount(assigned_experts, minlength=num_experts)

print("Tokens assigned to each expert (counting Top-k):", counts)

```

```python

all_counts = []

for i, logits in enumerate(router_logits):
    topk_scores, topk_indices = torch.topk(logits, k=model.config.top_k, dim=-1)

    assigned_experts = topk_indices.flatten()
    num_experts = logits.shape[-1]
    counts = torch.bincount(assigned_experts, minlength=num_experts)

    all_counts.append(counts)

    print(f"Layer {i}:")
    print("Tokens assigned to each expert:", counts)
    print("Total tokens:", counts.sum().item())
    print("-" * 40)
```

```python
import matplotlib.pyplot as plt

for i, counts in enumerate(all_counts):
    plt.figure(figsize=(10,4))
    plt.bar(range(len(counts)), counts.cpu().numpy())
    plt.xlabel("Expert index")
    plt.ylabel("Number of tokens assigned")
    plt.title(f"Layer {i} Expert Routing Histogram")
    plt.show()
```

```python
for i, counts in enumerate(all_counts):
    mean = counts.float().mean().item()
    std = counts.float().std().item()
    cv = std / mean
    print(f"Layer {i}: Mean={mean:.2f}, Std={std:.2f}, Coefficient of Variation={cv:.2f}")
```

```python


num_experts = router_logits[0].shape[-1]
total_counts = torch.zeros(num_experts, dtype=torch.int32, device="cpu")

for i, logits in enumerate(router_logits):
    logits_cpu = logits.cpu()

    topk_scores, topk_indices = torch.topk(logits_cpu, k=model.config.top_k, dim=-1)
    assigned_experts = topk_indices.flatten()

    counts = torch.bincount(assigned_experts, minlength=num_experts)

    total_counts += counts


total_counts_np = total_counts.numpy()


import matplotlib.pyplot as plt

plt.figure(figsize=(15,6))
plt.bar(range(num_experts), total_counts_np)
plt.xlabel("Expert Index")
plt.ylabel("Total Tokens Assigned Across All Layers")
plt.title("Cumulative Expert Token Assignment")
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```
